```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Positive skew (right)

```python
# Set seed for reproducibility
np.random.seed(42)
```

```python
# Example 1: Income Distribution
income_data = np.concatenate([
    np.random.normal(50000, 10000, 800),   # Moderate income earners
    np.random.normal(150000, 50000, 200)    # High-income earners
])

# Example 2: Home Prices in a Neighborhood
home_prices_data = np.concatenate([
    np.random.normal(300000, 50000, 800),   # Average home prices
    np.random.normal(800000, 100000, 200)   # Luxury home prices
])

# Example 3: Aging Population
age_data = np.concatenate([
    np.random.normal(70, 5, 800),   # Mostly elderly residents
    np.random.normal(90, 2, 200)    # Few much older individuals
])
```

```python
# Plotting
fig, axes = plt.subplots(3, 1, figsize=(8, 12))

# Example 1: Income Distribution
axes[0].hist(income_data, bins=30, density=True, color='blue', alpha=0.7)
axes[0].set_title('Income Distribution (Positively Skewed)')
axes[0].set_xlabel('Income')
axes[0].set_ylabel('Probability Density')

# Example 2: Home Prices in a Neighborhood
axes[1].hist(home_prices_data, bins=30, density=True, color='green', alpha=0.7)
axes[1].set_title('Home Prices Distribution (Positively Skewed)')
axes[1].set_xlabel('Home Prices')
axes[1].set_ylabel('Probability Density')

# Example 3: Aging Population
axes[2].hist(age_data, bins=30, density=True, color='orange', alpha=0.7)
axes[2].set_title('Age Distribution in a Retirement Community (Positively Skewed)')
axes[2].set_xlabel('Age')
axes[2].set_ylabel('Probability Density')

plt.tight_layout()
plt.show()
```

#### Negative skew (left)

```python
# Example 1: Exam Scores
exam_scores_data = np.concatenate([
    np.random.normal(80, 5, 800),   # High-scoring students
    np.random.normal(30, 10, 200)    # Few students with lower scores
])

# Example 2: Company Earnings
earnings_data = np.concatenate([
    np.random.normal(500000, 50000, 800),   # Companies with moderate profits
    np.random.normal(10000, 5000, 200)       # Few companies with significant losses
])

# Example 3: Rainfall in a Desert Region
# rainfall_data = np.concatenate([
#     np.random.normal(10, 5, 800),    # Minimal rainfall most of the time
#     np.random.normal(100, 20, 200)   # Occasional heavy rainstorms
# ])
# Generate rainfall data for a negatively skewed distribution
rainfall_data = np.concatenate([
    np.random.normal(100, 20, 800),   # Occasional heavy rainstorms
    np.random.normal(10, 5, 200)       # Minimal rainfall most of the time
])
```

```python
# Plotting
fig, axes = plt.subplots(3, 1, figsize=(8, 12))

# Example 1: Exam Scores
axes[0].hist(exam_scores_data, bins=30, density=True, color='blue', alpha=0.7)
axes[0].set_title('Exam Scores Distribution (Negatively Skewed)')
axes[0].set_xlabel('Exam Scores')
axes[0].set_ylabel('Probability Density')

# Example 2: Company Earnings
axes[1].hist(earnings_data, bins=30, density=True, color='green', alpha=0.7)
axes[1].set_title('Company Earnings Distribution (Negatively Skewed)')
axes[1].set_xlabel('Earnings')
axes[1].set_ylabel('Probability Density')

# Example 3: Rainfall in a Desert Region
axes[2].hist(rainfall_data, bins=30, density=True, color='orange', alpha=0.7)
axes[2].set_title('Rainfall Distribution in a Desert Region (Negatively Skewed)')
axes[2].set_xlabel('Rainfall')
axes[2].set_ylabel('Probability Density')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import skew
```

```python
# Positive Skewness (Right Skew) - Income Distribution
skew_income = skew(income_data)
print(f'Skewness for Income Distribution: {skew_income}')
```

```python
# Negative Skewness (Left Skew) - Exam Scores
skew_exam_scores = skew(exam_scores_data)
print(f'Skewness for Exam Scores: {skew_exam_scores}')
```

```python
# Positive Skewness (Right Skew) - Home Prices in a Neighborhood
skew_home_prices = skew(home_prices_data)
print(f'Skewness for Home Prices Distribution: {skew_home_prices}')
```

```python
# Negative Skewness (Left Skew) - Company Earnings
skew_earnings = skew(earnings_data)
print(f'Skewness for Company Earnings Distribution: {skew_earnings}')
```

```python
# Positive Skewness (Right Skew) - Aging Population
skew_age = skew(age_data)
print(f'Skewness for Age Distribution in a Retirement Community: {skew_age}')
```

```python
# Negative Skewness (Left Skew) - Rainfall in a Desert Region
skew_rainfall = skew(rainfall_data)
print(f'Skewness for Rainfall Distribution in a Desert Region: {skew_rainfall}')
```

```python
# Define skewness ranges and corresponding descriptions
ranges = [(-np.inf, -1), (-1, -0.5), (-0.5, 0.5), (0.5, 1), (1, np.inf)]
descriptions = ['Highly Skewed (Left)', 'Moderately Skewed (Left)', 'Fairly Symmetrical', 'Moderately Skewed (Right)', 'Highly Skewed (Right)']

# Generate data points for plotting
x_points = []
for r in ranges:
    x_points.extend([r[0], r[1]])

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x_points, [0] * len(x_points), marker='o', linestyle='', markersize=10, color='blue')

# Annotate skewness ranges
for i, desc in enumerate(descriptions):
    plt.text((ranges[i][0] + ranges[i][1]) / 2, 0.1, desc, ha='center', va='bottom', color='black', fontsize=10)

# Customize plot
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.title('Skewness Ranges and Descriptions')
plt.xlabel('Skewness')
plt.yticks([])  # Remove y-axis ticks
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```
