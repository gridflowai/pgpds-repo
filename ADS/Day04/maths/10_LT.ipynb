#### Examples of linear transformation

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Scaling**

```python
# Define the scaling transformation matrix
scaling_factor = 2
T = np.array([[scaling_factor, 0], 
              [0, scaling_factor]])
```

```python
T
```

```python
# Define the original vectors
vectors = np.array([[1, 1], 
                    [2, 2], 
                    [3, 1]])
```

```python
vectors
```

```python
# Apply the transformation
transformed_vectors = vectors @ T.T
```

```python
# Plotting
fig, ax = plt.subplots()
ax.quiver(*vectors.T, color='r', scale=1, scale_units='xy', angles='xy', label='Original')
ax.quiver(*transformed_vectors.T, color='b', scale=1, scale_units='xy', angles='xy', label='Transformed')
plt.xlim(0, 7)
plt.ylim(0, 7)
ax.set_aspect('equal')
plt.grid()
plt.legend()
plt.title("Scaling Transformation")
plt.show()
```

**Rotation**

```python
# Define the rotation transformation matrix
theta = np.pi / 4  # 45 degrees
T     = np.array([[np.cos(theta), -np.sin(theta)], 
                  [np.sin(theta), np.cos(theta)]])
```

```python
T
```

```python
# Define the original vectors
vectors = np.array([[1, 0], [0, 1], [1, 1]])
vectors
```

```python
# Apply the transformation
transformed_vectors = vectors @ T.T
```

```python
# Plotting
fig, ax = plt.subplots()
ax.quiver(*vectors.T, color='r', scale=1, scale_units='xy', angles='xy', label='Original')
ax.quiver(*transformed_vectors.T, color='b', scale=1, scale_units='xy', angles='xy', label='Transformed')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
ax.set_aspect('equal')
plt.grid()
plt.legend()
plt.title("Rotation Transformation")
plt.show()
```

**Shearing**

```python
# Define the shearing transformation matrix
shear_factor = 1
T = np.array([[1, shear_factor], [0, 1]])
T
```

```python
# Define the original vectors
vectors = np.array([[1, 1], [2, 2], [3, 1]])
vectors
```

```python
# Apply the transformation
transformed_vectors = vectors @ T.T
```

```python
# Plotting
fig, ax = plt.subplots()
ax.quiver(*vectors.T, color='r', scale=1, scale_units='xy', angles='xy', label='Original')
ax.quiver(*transformed_vectors.T, color='b', scale=1, scale_units='xy', angles='xy', label='Transformed')
plt.xlim(0, 7)
plt.ylim(0, 7)
ax.set_aspect('equal')
plt.grid()
plt.legend()
plt.title("Shearing Transformation")
plt.show()
```
