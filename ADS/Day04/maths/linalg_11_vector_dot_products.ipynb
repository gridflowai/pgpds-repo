### DOT PRODUCT
-------------------------

- In linear algebra, the dot product, also known as the scalar product or inner product, is a fundamental operation between two vectors that results in a scalar (a single numerical value) rather than another vector. 

- The dot product is often denoted using the symbol "·" or simply by placing two vectors side by side. Given two vectors, A and B, the dot product is denoted as either A · B or AB.

- The dot product of two vectors A and B is defined as:

    - A · B = |A| * |B| * cos(θ)

Where:

- A · B is the dot product of vectors A and B.
- |A| and |B| are the magnitudes (or lengths) of vectors A and B, respectively.
- θ (theta) is the angle between vectors A and B.

#### Properties

- `Commutativity`: A · B = B · A
    - The dot product is commutative, which means the order of the vectors does not matter.

- `Distributivity`: A · (B + C) = A · B + A · C
    - The dot product distributes over vector addition.

- `Linearity`: (aA) · B = a(A · B)
    - The dot product is linear with respect to scalar multiplication.

- `Orthogonal Vectors`: If the angle θ between two vectors is 90 degrees (π/2 radians), then their dot product is zero: A · B = 0. This indicates that the vectors are orthogonal (perpendicular) to each other.

- `Angle Relationship`: The dot product can be used to find the angle θ between two vectors:
    - cos(θ) = (A · B) / (|A| * |B|)

#### Applications of the dot product 

- Calculating the angle between two vectors.
- Determining whether vectors are orthogonal (perpendicular) to each other.
- Finding the projection of one vector onto another.
- Calculating work done by a force along a direction.
- Solving systems of linear equations using the dot product to form equations.

**Example 1: Calculating the Dot Product**

```python
import numpy as np
```

```python
# Define two vectors
vector_A = np.array([3, 4])
vector_B = np.array([1, 2])
```

```python
# Calculate the dot product
dot_product = np.dot(vector_A, vector_B)

print(f"The dot product of A and B is {dot_product}")
```

**Example 2: Finding the Angle Between Two Vectors**

```python
import matplotlib.pyplot as plt
```

```python
# Define two vectors
vector_A = np.array([3, 4])
vector_B = np.array([1, 2])
```

```python
# Calculate the dot product
dot_product = np.dot(vector_A, vector_B)
```

```python
# Calculate the magnitudes of the vectors
magnitude_A = np.linalg.norm(vector_A)
magnitude_B = np.linalg.norm(vector_B)
```

```python
# Calculate the angle between the vectors in degrees
angle_degrees = np.degrees(np.arccos(dot_product / (magnitude_A * magnitude_B)))

print(f"The angle between A and B is {angle_degrees} degrees")
```

```python
# Plot the vectors
plt.figure()
plt.quiver(0, 0, vector_A[0], vector_A[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector A')
plt.quiver(0, 0, vector_B[0], vector_B[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector B')
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.title('Vectors A and B')
```

**Example 3: Vector Projection Using Dot Product**


```python
# Define two vectors
vector_A = np.array([1, 2])
vector_B = np.array([3, 4])
```

```python
# Calculate the dot product
dot_product = np.dot(vector_A, vector_B)
```

```python
# Calculate the projection of A onto B
projection_A_on_B = (dot_product / np.linalg.norm(vector_B) ** 2) * vector_B
```

```python
# Plot the vectors and their projection
plt.figure()
plt.quiver(0, 0, vector_A[0], vector_A[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector A')
plt.quiver(0, 0, vector_B[0], vector_B[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector B')
plt.quiver(0, 0, projection_A_on_B[0], projection_A_on_B[1], angles='xy', scale_units='xy', scale=1, color='green', label='Projection of A on B')
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.title('Vector Projection')
```
