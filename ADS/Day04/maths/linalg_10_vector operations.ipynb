##### Example 1: Scaling a Vector

v=[2,3]. We'll scale it by a factor of c=3 to observe how the magnitude and direction change.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Original vector
v = np.array([2, 3])
```

```python
# Scalar multiplication
c = 3
scaled_vector = c * v
```

```python
# Plotting original and scaled vectors
plt.figure(figsize=(6, 4))
plt.axvline(x=0, color='gray', linewidth=0.5)
plt.axhline(y=0, color='gray', linewidth=0.5)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector')
plt.quiver(0, 0, scaled_vector[0], scaled_vector[1], angles='xy', scale_units='xy', scale=1, color='red', label='Scaled Vector')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scalar Multiplication: Scaling a Vector')
plt.legend(loc='upper left')
plt.grid()
```

##### Example 2: Scaling a Zero Vector
Let's see what happens when we scale the zero vector by any scalar c.

```python
# Zero vector
zero_vector = np.array([0, 0])
```

```python
# Scalar multiplication with different values of c
c_values = np.linspace(-10, 10, 20)
scaled_vectors = [c * zero_vector for c in c_values]
```

```python
# Plotting scaled zero vectors
plt.figure(figsize=(6, 6))
plt.axvline(x=0, color='gray', linewidth=0.5)
plt.axhline(y=0, color='gray', linewidth=0.5)
for scaled_vector in scaled_vectors:
    plt.quiver(0, 0, scaled_vector[0], scaled_vector[1], angles='xy', scale_units='xy', scale=1, color='blue')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scalar Multiplication: Scaling the Zero Vector')
plt.grid()
```

##### Example 1: Calculating Dot Product

Let's calculate the dot product of two vectors 

u=[3,4] and 
v=[5,2].

```python
# Example vectors
vector_u = np.array([3, 4])
vector_v = np.array([5, 2])

dot_product = np.dot(vector_u, vector_v)
print("Dot Product of u and v:", dot_product)
```

```python
# Plotting the vectors
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, vector_u[0], vector_u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u')
plt.quiver(0, 0, vector_v[0], vector_v[1], angles='xy', scale_units='xy', scale=1, color='red', label='v')

# Plotting the dot product as a dashed line
plt.plot([0, vector_u[0]], [0, vector_u[1]], 'b--', label='u')
plt.plot([0, vector_v[0]], [0, vector_v[1]], 'r--', label='v')
plt.text(1, 2, f'Dot Product: {dot_product}', fontsize=12, color='green')

# Set plot limits and labels
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.xlabel('x')
plt.ylabel('y')

# Add legend and grid
plt.legend()
plt.grid()
plt.title('Dot Product Visualization');
```

**Key points on dot products**
- 2 vectors
- based on "projection" of one vector onto the 2nd one
- we want to know `how much` of vector $a$ is pointing in the same direction in the vector $b$
- result
    - +ve
    - -ve 
    - 0

##### Cross Product (in 3D)

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define two 3D vectors
vector_a = np.array([1, 2, 3])
vector_b = np.array([4, 5, 6])
```

```python
# Calculate the cross product
cross_product = np.cross(vector_a, vector_b)
```

```python
# Create a figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the vectors
ax.quiver(0, 0, 0, vector_a[0], vector_a[1], vector_a[2], color='r', label='Vector A')
ax.quiver(0, 0, 0, vector_b[0], vector_b[1], vector_b[2], color='g', label='Vector B')
ax.quiver(0, 0, 0, cross_product[0], cross_product[1], cross_product[2], color='b', label='Cross Product')

# Set plot limits
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 10])

# Set labels and legend
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
```

The cross product is not a common operation in data science and machine learning compared to other vector operations like dot product, elementwise operations, and matrix multiplication. This is because the cross product is specifically defined for 3D vectors and has limited direct applications in the types of data and computations typically encountered in DS/ML.
