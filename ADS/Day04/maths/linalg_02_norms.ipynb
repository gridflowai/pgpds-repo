----------------------
#### Vector norms
----------------------------


```python
import numpy as np
import matplotlib.pyplot as plt

# Define a vector
vector = np.array([3, 4])

# Values of p for different norms
p_values = [1, 2, 3, 4, 5, np.inf]

# Create a figure and set axis limits
plt.figure(figsize=(10, 4))
plt.xlim(0, 16)
plt.ylim(0, 1)

# Plot the vector norms for each value of p using loops
for p in p_values:
    norm = np.linalg.norm(vector, ord=p)
    plt.plot([0, norm], [0, 0], marker='o', label=f'p={p}: {norm:.2f}')

# Add labels
plt.xlabel('Norm')
plt.ylabel('Dummy')
plt.ylim((-5, 5))
plt.title('Vector Norms Example for Different p Values')

# Add legend
plt.legend()

# Remove y-axis ticks and labels
plt.yticks([])

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random vectors around the origin with a constraint
num_vectors = 1000
vectors = np.random.uniform(-1, 1, size=(num_vectors, 2))
vectors = vectors / np.linalg.norm(vectors, ord=1, axis=1, keepdims=True)  # Normalize by L1 norm to satisfy the constraint

# Calculate L1 norms
norms_l1 = np.linalg.norm(vectors, ord=1, axis=1)

# Create a figure and set axis limits
plt.figure(figsize=(5, 5))
plt.xlim(-3.2, 3.2)
plt.ylim(-3.2, 3.2)

# Plot the vectors
plt.scatter(vectors[:, 0], vectors[:, 1], c=norms_l1, cmap='viridis', marker='o', s=30)

# Add labels
plt.xlabel('X')
plt.ylabel('Y')
plt.title('L1 Norms of Random Vectors with Constraint: Sum of Abs. Values = 1')

# Show the plot
plt.colorbar(label='$L_1$ Norm')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random vectors around the origin
num_vectors = 100
angles = np.linspace(0, 2 * np.pi, num_vectors)
vectors = np.array([np.cos(angles), np.sin(angles)]).T

# Calculate L2 norms
l2_norms = np.linalg.norm(vectors, ord=2, axis=1)

# Create a figure and set axis limits
plt.figure(figsize=(6, 6))
plt.xlim(-4.5, 4.5)
plt.ylim(-4.5, 4.5)

# Plot the L2 norms as points
plt.scatter(vectors[:, 0], vectors[:, 1], c=l2_norms, cmap='viridis', s=50)

# Add constraint line (sqrt(x^2 + y^2) = 1)
x_vals = np.linspace(-1.5, 1.5, 400)
y_vals_pos = np.sqrt(1 - x_vals ** 2)
y_vals_neg = -np.sqrt(1 - x_vals ** 2)
plt.plot(x_vals, y_vals_pos, color='red', linestyle='dashed')
plt.plot(x_vals, y_vals_neg, color='red', linestyle='dashed')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('L2 Norms of Random Vectors')
plt.colorbar(label='L2 Norm')

# Show the plot
plt.grid(True)
plt.axis('equal')
plt.show()
```
