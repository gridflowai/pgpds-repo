### Use cases : Linear transformation 

- Projections
------------------------------

- Projections are a common type of linear transformation in mathematics and computer graphics. 

- They involve mapping points onto a lower-dimensional subspace while preserving certain properties.

#### Orthogonal Projection onto a Line

- project a 2D point onto a line. Given a line defined by a unit vector v and a point p, we can compute the projection of p onto the line using the dot product.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the line onto which we want to project
line_direction = np.array([2, 1])  # Example: The line with direction [1, 1]
```

```python
# Define a set of vectors to project
vectors = np.array([[2, 3], [3, 1], [1, 2], [4, 4]])
```

```python
# Initialize an array to store the projections
projections = []

# Define colors for original vectors
vector_colors = ['r', 'c', 'k', 'yellow']
```

```python
# Calculate the projection of each vector onto the line
for vector in vectors:
    projection = np.dot(vector, line_direction) / np.dot(line_direction, line_direction) * line_direction
    projections.append(projection)

projections = np.array(projections)
```

```python
# First plot: Original vectors and projection line
plt.figure(figsize=(8, 4))

# Original vectors with specified colors
for i, (x, y) in enumerate(vectors):
    plt.quiver(x, y, angles='xy', scale_units='xy', scale=1, color=vector_colors[i], label=f'Original Vector {i+1}')
plt.quiver(0, 0, line_direction[0], line_direction[1], angles='xy', scale_units='xy', scale=1, color='green', label='Line Direction', width=0.02)
plt.title('Original Vectors with Projection Line')
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.grid()
plt.legend()
plt.show()

# Second plot: Projection line and projected vectors
plt.figure(figsize=(8, 4))

# Projection line
plt.quiver(0, 0, line_direction[0], line_direction[1], angles='xy', scale_units='xy', scale=1, color='green', label='Line Direction', width=0.02)

# Projected vectors with specified colors
for projection_color, projection in zip(vector_colors, projections):
    plt.quiver(projection[0], projection[1], angles='xy', scale_units='xy', scale=1, color=projection_color, label='Projected Vector')
plt.title('Projection Line with Projected Vectors')
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.grid()
plt.legend()
```
