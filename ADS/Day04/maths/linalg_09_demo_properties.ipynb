#### fundamental properties of vector operations
---------------------------

**Additive Property**

The `additive property` states that adding two vectors results in another vector. This property is central to various data science tasks involving combining or aggregating information.

- Example: Average Calculation

```python
import numpy as np

# Create an array of data points (vectors)
data = np.array([[3, 5],
                 [2, 4],
                 [4, 6]])

# Calculate the average (mean) vector
average_vector = np.mean(data, axis=0)

print("Data Points:\n", data)
print("\nAverage Vector:", average_vector)
```

In this example, the additive property is used to calculate the average vector by adding the vectors element-wise and then dividing by the number of vectors. This technique is useful for summarizing and visualizing data.

**Associative Property**

The associative property states that the order in which vectors are added does not affect the result. This property is often used in iterative algorithms and mathematical formulations.

- Example: K-Means Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate random data
np.random.seed(42)
data = np.random.randn(100, 2)

# Number of clusters
n_clusters = 3

# Initialize K-Means
kmeans = KMeans(n_clusters=n_clusters)

# Create subplots
plt.figure(figsize=(12, 4))
for i in range(1, n_clusters + 1):
    plt.subplot(1, n_clusters, i)

    # Fit K-Means with different initializations
    kmeans.n_init = i
    clusters = kmeans.fit_predict(data)

    # Plot the data points with different colors for each cluster
    for cluster_idx in range(n_clusters):
        plt.scatter(data[clusters == cluster_idx, 0], data[clusters == cluster_idx, 1], label=f'Cluster {cluster_idx}')

    plt.title(f'Initialization {i}')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()

plt.tight_layout()
plt.show()

```

n K-Means clustering, the associative property of vector addition is indirectly used. The algorithm iteratively assigns data points to clusters based on the Euclidean distance between data points and cluster centroids. The order in which data points are assigned does not affect the final cluster assignments.

**Scalar Multiplication**

Scalar multiplication involves scaling a vector by a scalar value. This operation is used to change the magnitude and direction of vectors.

- Example: Data Augmentation

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image_path = r'D:\AI-DATASETS\keras\datasets\img_aug\aria.JPG'  # Replace with your image path
image = cv2.imread(image_path)

# Define the brightness factor
brightness_factor = 50

# Perform image brightening by adding the brightness factor to all pixel values
brightened_image = cv2.add(image, brightness_factor)

# Convert BGR image to RGB for displaying with matplotlib
brightened_image_rgb = cv2.cvtColor(brightened_image, cv2.COLOR_BGR2RGB)

# Display the original and brightened images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(brightened_image_rgb)
plt.title('Brightened Image')

plt.tight_layout()
plt.show()
```

In this example, scalar multiplication is used to brighten an image by scaling the RGB values of the pixels. Scalar multiplication can be applied to augment data for machine learning models or perform various image processing tasks.
